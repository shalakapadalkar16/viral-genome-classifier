```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load splits
train_df = pd.read_csv('/Users/shalakapadalkar/Desktop/ML/Projects/viral-genome-classifier/data/splits/train.csv')
val_df = pd.read_csv('/Users/shalakapadalkar/Desktop/ML/Projects/viral-genome-classifier/data/splits/val.csv')
test_df = pd.read_csv('/Users/shalakapadalkar/Desktop/ML/Projects/viral-genome-classifier/data/splits/test.csv')

print(f"Train: {len(train_df)}")
print(f"Val: {len(val_df)}")
print(f"Test: {len(test_df)}")
```

```python
# Plot class distribution
fig, ax = plt.subplots(1, 3, figsize=(15, 4))
train_df['label'].value_counts().plot(kind='bar', ax=ax[0], title='Train')
val_df['label'].value_counts().plot(kind='bar', ax=ax[1], title='Val')
test_df['label'].value_counts().plot(kind='bar', ax=ax[2], title='Test')
plt.tight_layout()
plt.savefig('/Users/shalakapadalkar/Desktop/ML/Projects/viral-genome-classifier/results/figures/data_distribution.png')
```

```python
# Sequence length distribution
plt.figure(figsize=(10, 6))
train_df['length'].hist(bins=50)
plt.xlabel('Sequence Length (bp)')
plt.ylabel('Count')
plt.title('Sequence Length Distribution')
plt.savefig('/Users/shalakapadalkar/Desktop/ML/Projects/viral-genome-classifier/results/figures/length_distribution.png')
```

```python
# GC content distribution
plt.figure(figsize=(10, 6))
train_df['gc_content'].hist(bins=50)
plt.xlabel('GC Content')
plt.ylabel('Count')
plt.title('GC Content Distribution')
plt.savefig('/Users/shalakapadalkar/Desktop/ML/Projects/viral-genome-classifier/results/figures/gc_distribution.png')
```

```python
# Check for issues
print("Length statistics:")
print(train_df['length'].describe())

print("\nGC content statistics:")
print(train_df['gc_content'].describe())

# Check for extreme values
print("\nSequences with extreme GC:")
print(train_df[train_df['gc_content'] < 0.35].shape[0], "low GC")
print(train_df[train_df['gc_content'] > 0.65].shape[0], "high GC")
```
